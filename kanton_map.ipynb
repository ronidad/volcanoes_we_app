```python
import pandas as pd
import folium
print (folium.__version__)
```

```python
df = pd.read_csv('cod.csv')
df
lat = list(df['lat'])
lon = list(df['lon'])
nam = list(df['name'])
ren = list(df['rent'])
lon, lat, nam, ren
```

```python
#Display the map
def display_rent(rent):
    if rent <25000:
        return "green"
    else:
        return "blue"

kanton_map = folium.Map(location=[-1.360600, 36.991018], tiles= "Stamen Terrain", zoom_start=12)
#kanton_map
cordinates = [[-1.360600, 36.991018],[-1.3485974250766721, 36.981037220261754], [-1.3397573830948588, 36.957940872086624]]
for la, lo, n,r in zip(lat,lon, nam, ren):
    # for cod in cordinates:
    folium.CircleMarker(location=[la,lo], popup=n, radius=6, fill_color=display_rent(r), color = 'grey'  ).add_to(kanton_map)


# folium.Marker(location=[-1.360600, 36.991018], popup='Mosomi', icon=folium.Icon(color='green')).add_to(kanton_map)
# folium.Marker(location=[-1.3485974250766721, 36.981037220261754], popup='Jabin', icon=folium.Icon(color='green')).add_to(kanton_map)
# 
kanton_map
```
